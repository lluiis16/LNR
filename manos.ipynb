```python
import json
import cv2
import base64
import numpy as np
import os
from sklearn.feature_extraction import image
from sklearn.cluster import MiniBatchKMeans

pathJson = "C:\\Users\\Lluis\\OneDrive\\Escritorio\\3r Curso\\MDP II\\trabajo\\HANDS" # Change to the folder where the json are located
jsonFiles = [fJson for fJson in os.listdir(pathJson) if fJson.endswith(".json")]
# print(jsonFiles)
pSize = (5, 5)
bufferData = []
kmeans = MiniBatchKMeans(n_clusters=81, random_state=0, verbose=True)
Y=[]
for fJ in jsonFiles:
    with open(pathJson+"/"+fJ) as infile:
        data = json.load(infile)
        for item in data["shapes"]:
            label = item["label"]; Y.append(label); points = item["points"]
            imdata = base64.b64decode(data["imageData"])
            npimg = np.frombuffer(imdata, dtype=np.uint8);
            im0 = cv2.imdecode(npimg, 1)
            im0 = cv2.resize(im0,(150,150)) # image size is normalized
        patches = image.extract_patches_2d(im0, pSize, max_patches=50, random_state=0)
        patches = np.reshape(patches, (len(patches), -1)) # max_patches x (pSize x pSize x channels)
        bufferData.append(patches)
        
dataIm = np.concatenate(bufferData, axis=0)
dataIm = dataIm.astype(float)
dataIm -= np.mean(dataIm, axis=0)
dataIm /= np.std(dataIm, axis=0)
kmeans.partial_fit(dataIm)

import matplotlib.pyplot as plt # plot centroids to have a visual view

plt.figure(figsize=(4.2, 4))

for i, patch in enumerate(kmeans.cluster_centers_):
    plt.subplot(9, 9, i + 1)
    plt.imshow(patch.reshape((5,5,3)), cmap=plt.cm.gray, interpolation="nearest")
    plt.xticks(()); plt.yticks(())
plt.suptitle(
    "Patches",
    fontsize=16,
)

plt.subplots_adjust(0.08, 0.02, 0.92, 0.85, 0.08, 0.23)
plt.show()
```
